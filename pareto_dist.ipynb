```python
import re
from operator import itemgetter
import matplotlib.pyplot as plt
from scipy import special
import numpy as np
import seaborn as sns
from matplotlib.pyplot import figure
```

Создаем списки с объемами корпусов и названиями размеров

```python
values = [19309336, 10016740, 7513256, 6313715, 3220555, 2494267, 2243986, 1867995, 1635043, 1624266, 
          1350636, 1235530, 1217859, 1111264, 774761, 704576, 479904, 429938, 407569, 324076,
          268616, 199543, 157662, 154841]
meters = 'Я4 Я5 Х4 Ан3 Х5 Ам4 НУР Ан4 Я3 Ам3 Д4 Я6 Ан2 Ан5 Х6 Х3 Д3 Ам5 Х8 Д5 Я7 Ам2 Я2 Я8'
meters = meters.split(' ')
```

Создаем список со значениями, распределенными по степенному закону, за основу берем объем самого большого корпуса

```python
pareto = []
for i in range(len(values)):
    if i == 0:
        pareto.append(values[0])
    else:
        pareto.append(int(values[0] // (i+1)))
len(pareto)
```

Определяем параметры визуализации

```python
large = 22; med = 16; small = 12
params = {'axes.titlesize': large,
          'legend.fontsize': med,
          'figure.figsize': (16, 10),
          'axes.labelsize': med,
          'axes.titlesize': med,
          'xtick.labelsize': med,
          'ytick.labelsize': med,
          'figure.titlesize': large}
plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style("white")
sns.set_theme()
sns.set_color_codes()
%matplotlib inline
```

Строим гистограммы для распределения Ципфа и реального распределения объемов корпусов

```python
X = np.array(meters)
pareto = np.array(pareto)
Y = np.array(values)
plt.figure(figsize=(12,10), dpi= 80)
sns.barplot(X, Y, color='royalblue', label='Объем корпуса') 
sns.lineplot(X, pareto, color='salmon', label='Распределение Парето')

plt.legend(loc='upper right', prop={'size': 15})
plt.show()
```
